```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load("roomba_training_data.npz")

rewards = data['rewards']
lengths = data['lengths']
cells = data['cells']
collisions = data['collisions']
coverage = data['coverage']

episodes = np.arange(1, len(rewards)+1)

def moving_average(x, window=10):
    return np.convolve(x, np.ones(window)/window, mode='valid')

# episode rewards
plt.figure(figsize=(10,5))
plt.plot(episodes, rewards, label='Reward per episode')
plt.plot(episodes[len(episodes)-len(moving_average(rewards)):], moving_average(rewards), label='Reward MA (window=10)')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Episode Rewards')
plt.legend()
plt.grid(True)
plt.show()

# episode lengths
plt.figure(figsize=(10,5))
plt.plot(episodes, lengths, label='Episode Length')
plt.plot(episodes[len(episodes)-len(moving_average(lengths)):], moving_average(lengths), label='Length MA (window=10)')
plt.xlabel('Episode')
plt.ylabel('Length')
plt.title('Episode Lengths')
plt.legend()
plt.grid(True)
plt.show()

# new cells per episode 
plt.figure(figsize=(10,5))
plt.plot(episodes, cells, label='New Cells per Episode')
plt.plot(episodes[len(episodes)-len(moving_average(cells)):], moving_average(cells), label='Cells MA (window=10)')
plt.xlabel('Episode')
plt.ylabel('New Cells')
plt.title('Exploration: New Cells per Episode')
plt.legend()
plt.grid(True)
plt.show()

# Plot collisions per episode
plt.figure(figsize=(10,5))
plt.plot(episodes, collisions, label='Collisions per Episode')
plt.plot(episodes[len(episodes)-len(moving_average(collisions)):], moving_average(collisions), label='Collisions MA (window=10)')
plt.xlabel('Episode')
plt.ylabel('Collisions')
plt.title('Collisions per Episode')
plt.legend()
plt.grid(True)
plt.show()

# coverage per episode
plt.figure(figsize=(10,5))
plt.plot(episodes, coverage, label='Coverage per Episode')
plt.plot(episodes[len(episodes)-len(moving_average(coverage)):], moving_average(coverage), label='Coverage MA (window=10)')
plt.xlabel('Episode')
plt.ylabel('Coverage')
plt.title('Room Coverage per Episode')
plt.legend()
plt.grid(True)
plt.show()
```
